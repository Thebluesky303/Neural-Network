# 1. Importing the libraries and dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

# from google.colab.patches import cv2_imshow 
from PIL import Image
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.math import confusion_matrix
```

# 2. Loading the MNIST dataset from keras

```python
from keras.datasets import mnist
(X_train, y_train),(X_test, y_test) = mnist.load_data()
```

```python
type(X_train)
```

```python
#shape of the numpy array
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
X_train.shape
```

```python
#displaying the image
plt.imshow(X_train[25])
plt.show()

print(y_train[50])
```

```python
#unique values in y_train
print(np.unique(y_train))
print(np.unique(y_test))
```

We can use this labels as such or we can use also apply one hot encoding 

if the datset dimension is not equal dimension we can used we have to convert to same dimension

```python
#scalling the values
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[10]
```

# 3. Building the neural network

```python
# setting up the layers of the neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(50,activation ='relu'),
    keras.layers.Dense(50, activation= 'relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer ='adam',
             loss = 'sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#trainig the neural network
model.fit(X_train,y_train, epochs=10)
```

# 4. Model Evalution

```python
loss, accuracy = model.evaluate(X_test,y_test)
print(accuracy)
```

```python
# first data of test data
plt.imshow(X_test[0])
plt.show()
```

```python
print(y_test[0])
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred.shape)
```

```python
print(y_pred[0])
```

```python
# converting the prediction prob to label
label_for_fist_image = np.argmax(y_pred[0])
print(label_for_fist_image)
```

```python
# converting the prediction prob to label for all test data points
y_pred_label = [np.argmax(i) for i in y_pred]
print(y_pred_label)
```

```python
print(y_pred)
```

# 5. Confusion matrix

```python
conf_mat = confusion_matrix(y_test, y_pred_label)
```

```python
print(conf_mat)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(conf_mat, annot = True, fmt = 'd', cmap= 'Blues')
plt.ylabel('True labels')
plt.xlabel('Predicted label')
```
