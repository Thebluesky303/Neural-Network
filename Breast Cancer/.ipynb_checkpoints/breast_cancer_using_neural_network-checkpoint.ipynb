# Problem Statement

The objective of this project is to build a machine learning model that can accurately classify breast tumors as malignant or benign based on various cell nucleus measurements. The dataset contains diagnostic features such as radius, texture, perimeter, area, smoothness, compactness, concavity, symmetry, and fractal dimension, computed from digitized images of breast tissue samples. By analyzing these features, the model aims to assist in early and accurate breast cancer diagnosis, thereby supporting clinical decision-making.

# Objectives

1. To analyze breast cancer diagnostic data containing cell nucleus measurements
2. To preprocess and transform the data for effective model training
3. To build a simple neural network model to classify tumors as malignant or benign
4. To evaluate the model’s performance using appropriate metrics
5. To support early and accurate breast cancer diagnosis through predictive modeling

# Using sklearn build in dataset

```python
data = sklearn.datasets.load_breast_cancer()
```

# 4. Split Data

```python
X = df.drop(columns='label', axis=1)
y = df['label']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

# 5. Building the Neural Network

## 5.1 Simple one hidden layer neural network

```python
#setting up the layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation= 'relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer ='adam',
             loss = 'sparse_categorical_crossentropy',
             metrics =['accuracy'])
```

```python
#training the neural network
history = model.fit(X_train_std, y_train, validation_split=0.1, epochs=10)
```

```python
#visualization of accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc= 'lower right')
```

```python
#visualization of loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc= 'upper right')
```

## 5.2 Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
print(accuracy)
```

## 5.4 Building the predicting system


```python
input_data = (20.29, 14.34, 135.1, 1297.0, 0.1003, 0.1328, 0.198, 0.1043, 0.1809, 0.05883, 0.7572, 0.7813, 5.438, 94.44, 0.01149, 0.02461, 0.05688, 0.01885, 0.01756, 0.005115, 22.54, 16.67, 152.2, 1575.0, 0.1374, 0.205, 0.4, 0.1625, 0.2364, 0.07678)
#change the input dat to a numpy array
input_data_as_numpy_array = np.asarray(input_data)
#reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
#standardlizing the data
input_data_std  = scaler.transform(input_data_reshaped)
prediction = model.predict(input_data_std)
prediction_label=[np.argmax(prediction)]
if(prediction_label[0]==0):
    print('The tumor is Malignant.')
else:
    print('The tumor is Benign')
```
