#Customer Churn Prediction Using Artificial Neural Network (ANN)

The Telco Customer Churn dataset contains customer-level information from a telecom company and is commonly used to study why customers leave a service. Each row represents one customer, while the columns describe their profile, subscribed services, account details, and churn status. It includes whether the customer left in the last month, the types of services they use, billing and contract information, and basic demographic details such as gender, senior citizen status, partner, and dependents. The dataset is useful for churn analysis, customer retention studies, and predictive modelling.

## printing the yes and no columns.

```python
def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

## replacing no internet service or no plone service, replaced by no

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

#5. convert yes and no to 1 and 0

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup', 'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.gender.unique()
```

#6. One hot encoding for categorical columns

```python
df2 = pd.get_dummies(
    data=df1,
    columns=['InternetService', 'Contract', 'PaymentMethod'],
    dtype=int
)
```

```python
df2.sample(5)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
  print(f'{col}: {df2[col].unique()}')
```

#7. Train test split

```python
x = df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape, x_test.shape
```

```python
len(x_train.columns)
```

#8. Build a model (ANN) in tensorflow/keras

```python
model = keras.Sequential([
    keras.layers.Dense(26,input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=200)
```

#9. model evaluation

```python
model.evaluate(x_test, y_test)
```

```python
yp = model.predict(x_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
  if element > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
